```python
from ipywidgets import interact, interactive, fixed, interact_manual, Text, Button, Output
import ipywidgets as widgets
from IPython.display import display
from src.chat import answer, create_rag_chain
```

```python
rag_chain = create_rag_chain()
```

```python
text_input = Text(value='Ask any question...')
button = Button(description='Generate')
out = Output()

def generate(b):
    out.clear_output()
    button.disabled=True
    prompt = text_input.value
    rep, question = answer(prompt, rag_chain)
    ans = rep[rep.find("\nCâu trả lời:")+len("\nCâu trả lời:"):].strip()
    with out:
        print(ans)
    button.disabled=False

button.on_click(generate)
display(text_input, button, out)
```
